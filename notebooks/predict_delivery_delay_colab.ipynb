```python
# -*- coding: utf-8 -*-
"""Previsão de Atrasos na Entrega - Modelo Completo.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1abc123
"""

# 0) INSTALAÇÕES E IMPORTS
print(" Instalando e importando bibliotecas...")

!pip install -q xgboost imbalanced-learn

import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (roc_auc_score, roc_curve, accuracy_score,
                           precision_score, recall_score, f1_score,
                           confusion_matrix, classification_report)
from imblearn.over_sampling import SMOTE
import joblib
import warnings
warnings.filterwarnings('ignore')

print(" Bibliotecas importadas com sucesso!")

# 1) CARREGAMENTO DE DADOS
print("\n Carregando dados...")

from google.colab import files

DATA_FILES = [
    'fato_entregas.csv',
    'dim_clientes.csv',
    'dim_transportadoras.csv',
    'dim_cidades.csv',
    'dim_calendario.csv',
    'metricas_agregadas.csv'
]

# Verificar arquivos locais primeiro
missing = [f for f in DATA_FILES if not Path(f).exists()]
if missing:
    print(' Arquivos não encontrados localmente:', missing)
    print(' Por favor faça upload dos arquivos agora...')
    uploaded = files.upload()
    for fn in uploaded.keys():
        print(f' Uploaded: {fn}')
else:
    print(' Todos os arquivos encontrados localmente!')

# Função segura para leitura de CSV
def safe_read_csv(file_path):
    """Carrega CSV com tratamento de erros"""
    try:
        df = pd.read_csv(file_path, low_memory=False)
        print(f" {file_path} carregado - Shape: {df.shape}")
        return df
    except Exception as e:
        print(f' Erro ao ler {file_path}: {e}')
        return None

# Carregar todos os datasets
fato = safe_read_csv('fato_entregas.csv')
dim_clientes = safe_read_csv('dim_clientes.csv')
dim_transportadoras = safe_read_csv('dim_transportadoras.csv')
dim_cidades = safe_read_csv('dim_cidades.csv')
dim_calendario = safe_read_csv('dim_calendario.csv')
metricas_agregadas = safe_read_csv('metricas_agregadas.csv')

# Verificação de carregamento
print("\n Resumo dos datasets carregados:")
datasets = {
    'fato_entregas': fato,
    'dim_clientes': dim_clientes,
    'dim_transportadoras': dim_transportadoras,
    'dim_cidades': dim_cidades,
    'dim_calendario': dim_calendario,
    'metricas_agregadas': metricas_agregadas
}

for name, df in datasets.items():
    if df is not None:
        print(f" {name}: {df.shape}")
        print(f"   Colunas: {list(df.columns)}")
    else:
        print(f" {name}: Não carregado")

# 2) ANÁLISE EXPLORATÓRIA DOS DADOS (EDA)
print("\n Análise Exploratória dos Dados...")

if fato is not None:
    print("\n Primeiras linhas do fato_entregas:")
    display(fato.head(3))

    print("\n Informações do dataset:")
    print(f"Shape: {fato.shape}")

    print("\n Tipos de dados:")
    print(fato.dtypes)

    print("\n Estatísticas descritivas:")
    display(fato.describe())

    print("\n Valores missing:")
    missing_data = fato.isna().sum()
    for col, count in missing_data[missing_data > 0].items():
        print(f"  {col}: {count} ({count/len(fato)*100:.1f}%)")

    if not missing_data.any():
        print("   Nenhum valor missing encontrado")

    # Verificar target
    target_col = 'is_delayed'
    if target_col in fato.columns:
        print(f"\n Variável target encontrada: '{target_col}'")
        target_dist = fato[target_col].value_counts()
        print("Distribuição:")
        print(target_dist)
        print("\nProporções:")
        print(target_dist / len(fato))

        # Plot distribuição do target
        plt.figure(figsize=(10, 5))

        plt.subplot(1, 2, 1)
        target_dist.plot(kind='bar', color=['skyblue', 'salmon'])
        plt.title('Distribuição do Target')
        plt.xlabel(target_col)
        plt.ylabel('Contagem')
        plt.xticks(rotation=0)

        plt.subplot(1, 2, 2)
        plt.pie(target_dist.values, labels=['Não Atrasado', 'Atrasado'],
                autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
        plt.title('Proporção do Target')

        plt.tight_layout()
        plt.show()
    else:
        print(f" Coluna target '{target_col}' não encontrada!")
        print("Colunas disponíveis:", list(fato.columns))

else:
    print(" fato_entregas não carregado!")

# 3) FEATURE ENGINEERING SIMPLIFICADO
print("\n Feature Engineering...")

if fato is not None and 'is_delayed' in fato.columns:
    # Criar cópia para não modificar o original
    df_processed = fato.copy()

    # CONVERSÃO DE DATAS
    date_columns = ['order_date', 'estimated_delivery', 'actual_delivery']
    for col in date_columns:
        if col in df_processed.columns:
            df_processed[col] = pd.to_datetime(df_processed[col], errors='coerce')

    # FEATURES TEMPORAIS da order_date
    if 'order_date' in df_processed.columns:
        df_processed['order_date'] = pd.to_datetime(df_processed['order_date'], errors='coerce')
        df_processed['year'] = df_processed['order_date'].dt.year
        df_processed['month'] = df_processed['order_date'].dt.month
        df_processed['day'] = df_processed['order_date'].dt.day
        df_processed['dayofweek'] = df_processed['order_date'].dt.dayofweek
        df_processed['is_weekend'] = df_processed['dayofweek'].isin([5, 6]).astype(int)
        df_processed['quarter'] = df_processed['order_date'].dt.quarter

    # FEATURES DE DURAÇÃO (se temos datas de entrega)
    if all(col in df_processed.columns for col in ['order_date', 'actual_delivery']):
        df_processed['delivery_duration_days'] = (df_processed['actual_delivery'] - df_processed['order_date']).dt.days

    # FEATURES CATEGÓRICAS - criar variáveis dummy para carrier
    if 'carrier' in df_processed.columns:
        carrier_dummies = pd.get_dummies(df_processed['carrier'], prefix='carrier')
        df_processed = pd.concat([df_processed, carrier_dummies], axis=1)

    # FEATURES DE LOCALIZAÇÃO - criar dummies para cidades
    for col in ['origin_city', 'destination_city']:
        if col in df_processed.columns:
            city_dummies = pd.get_dummies(df_processed[col], prefix=col)
            df_processed = pd.concat([df_processed, city_dummies], axis=1)

    # INTEGRAÇÃO COM CLIENTES (se disponível)
    if dim_clientes is not None and 'customer_id' in df_processed.columns:
        # Verificar colunas disponíveis
        client_cols_available = [col for col in dim_clientes.columns if col != 'customer_id']
        if client_cols_available:
            df_processed = df_processed.merge(
                dim_clientes[['customer_id'] + client_cols_available],
                on='customer_id', how='left'
            )
            print(f" Integradas {len(client_cols_available)} colunas de clientes")

    # FEATURES AGRREGADAS
    print(" Criando features agregadas...")

    # Por cliente
    if 'customer_id' in df_processed.columns:
        cust_agg = df_processed.groupby('customer_id').agg({
            'is_delayed': ['mean', 'count']
        })
        cust_agg.columns = ['cust_delay_rate', 'cust_total_orders']
        df_processed = df_processed.merge(cust_agg, on='customer_id', how='left')
        print(" Features agregadas por cliente criadas")

    # Por transportadora
    if 'carrier' in df_processed.columns:
        carrier_agg = df_processed.groupby('carrier').agg({
            'is_delayed': ['mean', 'count']
        })
        carrier_agg.columns = ['carrier_delay_rate', 'carrier_total_orders']
        df_processed = df_processed.merge(carrier_agg, on='carrier', how='left')
        print(" Features agregadas por transportadora criadas")

    print(f"\n Feature Engineering concluído!")
    print(f" Dataset final: {df_processed.shape}")

else:
    print(" Dados não disponíveis para processamento!")
    # Criar dataset básico para demonstração
    if fato is not None:
        df_processed = fato.copy()
        if 'is_delayed' not in df_processed.columns:
            df_processed['is_delayed'] = 0

# 4) PREPARAÇÃO PARA MODELAGEM
print("\n Preparando dados para modelagem...")

if 'df_processed' in locals() and df_processed is not None:
    # SELECIONAR FEATURES NUMÉRICAS E CATEGÓRICAS
    exclude_cols = ['delivery_id', 'customer_id', 'order_date', 'estimated_delivery',
                   'actual_delivery', 'status', 'delay_category', 'order_day_of_week']

    # Remover colunas de identificação e datas
    feature_cols = [col for col in df_processed.columns
                   if col not in exclude_cols + ['is_delayed']
                   and df_processed[col].dtype in [np.number, bool]]

    print(f" Usando {len(feature_cols)} features numéricas")
    print("Features:", feature_cols)

    # Criar dataset final
    X = df_processed[feature_cols].copy()
    y = df_processed['is_delayed']

    # Preencher valores missing
    for col in X.columns:
        if X[col].isna().any():
            if X[col].dtype in [np.float64, np.int64]:
                X[col] = X[col].fillna(X[col].median())
            else:
                X[col] = X[col].fillna(X[col].mode()[0] if len(X[col].mode()) > 0 else 0)

    print(f" Dataset preparado:")
    print(f"  X: {X.shape}")
    print(f"  y: {y.shape} (média: {y.mean():.3f})")

else:
    print(" Dados não disponíveis para modelagem!")

# 5) DIVISÃO TREINO/TESTE
print("\n Divisão dos dados...")

if 'X' in locals() and 'y' in locals():
    # Divisão estratificada para manter proporção do target
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    print(f" Divisão concluída:")
    print(f"  X_train: {X_train.shape}")
    print(f"  X_test:  {X_test.shape}")
    print(f"  y_train: {y_train.shape} ({y_train.mean():.1%} atrasados)")
    print(f"  y_test:  {y_test.shape} ({y_test.mean():.1%} atrasados)")

else:
    print(" Dados não disponíveis para divisão!")

# 6) TREINAMENTO DOS MODELOS
print("\n Treinando modelos...")

if 'X_train' in locals() and X_train is not None:
    results = {}

    # VERIFICAR BALANCEAMENTO
    print(" Verificando balanceamento...")
    class_ratio = y_train.value_counts(normalize=True)
    print(f"  Distribuição: Não Atrasado {class_ratio[0]:.3f} vs Atrasado {class_ratio[1]:.3f}")

    # Aplicar SMOTE se necessário
    if class_ratio.min() < 0.3:
        print("  Aplicando SMOTE para balanceamento...")
        smote = SMOTE(random_state=42)
        X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
        print(f"  Antes: {X_train.shape} → Depois: {X_train_balanced.shape}")
        X_train_used, y_train_used = X_train_balanced, y_train_balanced
    else:
        print(" Dados balanceados, usando dados originais")
        X_train_used, y_train_used = X_train, y_train

    # 6.1) REGRESSÃO LOGÍSTICA
    print("\n 6.1) Treinando Regressão Logística...")
    try:
        pipe_lr = Pipeline([
            ('scaler', StandardScaler()),
            ('clf', LogisticRegression(
                max_iter=1000,
                class_weight='balanced',
                random_state=42
            ))
        ])

        pipe_lr.fit(X_train_used, y_train_used)
        y_pred_lr = pipe_lr.predict(X_test)
        y_proba_lr = pipe_lr.predict_proba(X_test)[:, 1]

        results['logistic'] = {
            'model': pipe_lr,
            'y_pred': y_pred_lr,
            'y_proba': y_proba_lr
        }
        print(" Regressão Logística treinada!")
    except Exception as e:
        print(f" Erro na Regressão Logística: {e}")

    # 6.2) RANDOM FOREST
    print("\n 6.2) Treinando Random Forest...")
    try:
        rf_model = RandomForestClassifier(
            n_estimators=100,
            max_depth=10,
            class_weight='balanced',
            random_state=42,
            n_jobs=-1
        )

        rf_model.fit(X_train_used, y_train_used)
        y_pred_rf = rf_model.predict(X_test)
        y_proba_rf = rf_model.predict_proba(X_test)[:, 1]

        results['random_forest'] = {
            'model': rf_model,
            'y_pred': y_pred_rf,
            'y_proba': y_proba_rf
        }
        print(" Random Forest treinado!")
    except Exception as e:
        print(f" Erro no Random Forest: {e}")

    # 6.3) XGBOOST
    print("\n 6.3) Treinando XGBoost...")
    try:
        import xgboost as xgb

        xgb_model = xgb.XGBClassifier(
            n_estimators=100,
            max_depth=6,
            learning_rate=0.1,
            eval_metric='logloss',
            use_label_encoder=False,
            random_state=42
        )

        xgb_model.fit(X_train_used, y_train_used)
        y_pred_xgb = xgb_model.predict(X_test)
        y_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]

        results['xgboost'] = {
            'model': xgb_model,
            'y_pred': y_pred_xgb,
            'y_proba': y_proba_xgb
        }
        print(" XGBoost treinado!")
    except Exception as e:
        print(f" Erro no XGBoost: {e}")

    print(f"\n Modelos treinados: {list(results.keys())}")

else:
    print(" Dados de treino não disponíveis!")

# 7) AVALIAÇÃO DOS MODELOS
print("\n Avaliando modelos...")

def evaluate_model(name, y_true, y_pred, y_proba):
    """Avalia modelo com múltiplas métricas"""

    auc = roc_auc_score(y_true, y_proba)
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    cm = confusion_matrix(y_true, y_pred)

    return {
        'name': name,
        'auc': auc,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'confusion_matrix': cm
    }

if 'results' in locals() and results:
    all_metrics = []

    for model_name, result in results.items():
        print(f"\n{'='*50}")
        print(f" {model_name.upper()}")
        print(f"{'='*50}")

        metrics = evaluate_model(model_name, y_test, result['y_pred'], result['y_proba'])
        all_metrics.append(metrics)

        print(f"ROC AUC:       {metrics['auc']:.4f}")
        print(f"Acurácia:      {metrics['accuracy']:.4f}")
        print(f"Precisão:      {metrics['precision']:.4f}")
        print(f"Recall:        {metrics['recall']:.4f}")
        print(f"F1-Score:      {metrics['f1']:.4f}")

        print(f"\n Matriz de Confusão:")
        print(metrics['confusion_matrix'])

        # Curva ROC
        fpr, tpr, _ = roc_curve(y_test, result['y_proba'])

        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, label=f'{model_name} (AUC = {metrics["auc"]:.3f})', linewidth=2)
        plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'Curva ROC - {model_name}')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

    # COMPARAÇÃO ENTRE MODELOS
    print(f"\n{'='*60}")
    print(" COMPARAÇÃO ENTRE MODELOS")
    print(f"{'='*60}")

    metrics_df = pd.DataFrame(all_metrics).set_index('name')

    # Plot comparação
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    metrics_df['auc'].plot(kind='bar', ax=axes[0,0], color='skyblue', alpha=0.7)
    axes[0,0].set_title('ROC AUC Score')
    axes[0,0].set_ylim(0, 1)

    metrics_df['f1'].plot(kind='bar', ax=axes[0,1], color='lightgreen', alpha=0.7)
    axes[0,1].set_title('F1-Score')
    axes[0,1].set_ylim(0, 1)

    metrics_df[['precision', 'recall']].plot(kind='bar', ax=axes[1,0], alpha=0.7)
    axes[1,0].set_title('Precision vs Recall')
    axes[1,0].set_ylim(0, 1)

    metrics_df['accuracy'].plot(kind='bar', ax=axes[1,1], color='orange', alpha=0.7)
    axes[1,1].set_title('Accuracy')
    axes[1,1].set_ylim(0, 1)

    plt.tight_layout()
    plt.show()

    # Identificar melhor modelo
    best_model_name = metrics_df['auc'].idxmax()
    best_auc = metrics_df.loc[best_model_name, 'auc']
    best_metric = [m for m in all_metrics if m['name'] == best_model_name][0]

    print(f"\n MELHOR MODELO: {best_model_name}")
    print(f" Melhor AUC: {best_auc:.4f}")

    # Ranking
    print(f"\n RANKING (por AUC):")
    ranking = metrics_df['auc'].sort_values(ascending=False)
    for i, (model, score) in enumerate(ranking.items(), 1):
        print(f"  {i}º {model}: {score:.4f}")

else:
    print(" Nenhum modelo para avaliar!")

# 8) SALVANDO ARTEFATOS
print("\n Salvando artefatos...")

try:
    if 'results' in locals() and results and 'best_model_name' in locals():
        best_model = results[best_model_name]['model']

        print(f" Salvando melhor modelo: {best_model_name}")

        # Salvar modelo
        model_filename = f'melhor_modelo_{best_model_name}.pkl'
        joblib.dump(best_model, model_filename)
        print(f" Modelo salvo: {model_filename}")

        # Salvar feature names
        feature_names = X.columns.tolist()
        joblib.dump(feature_names, 'feature_names.pkl')
        print(" Feature names salvo")

        # Salvar métricas
        metrics_df = pd.DataFrame(all_metrics)
        metrics_df.to_csv('metricas_modelos.csv', index=False)
        print(" Métricas salvas")

        # Salvar previsões
        best_predictions = pd.DataFrame({
            'y_true': y_test.values,
            'y_pred': results[best_model_name]['y_pred'],
            'y_proba': results[best_model_name]['y_proba']
        })
        best_predictions.to_csv('previsoes_melhor_modelo.csv', index=False)
        print(" Previsões salvas")

        print(f"\n Resumo do melhor modelo:")
        print(f"  AUC: {best_metric['auc']:.4f}")
        print(f"  F1-Score: {best_metric['f1']:.4f}")
        print(f"  Acurácia: {best_metric['accuracy']:.4f}")

except Exception as e:
    print(f" Erro ao salvar artefatos: {e}")

# 9) DOWNLOAD DOS ARQUIVOS
print("\n Download dos arquivos...")

try:
    from google.colab import files

    files_to_download = [
        'melhor_modelo_logistic.pkl', 'melhor_modelo_random_forest.pkl', 'melhor_modelo_xgboost.pkl',
        'feature_names.pkl', 'metricas_modelos.csv', 'previsoes_melhor_modelo.csv'
    ]

    downloaded = 0
    for file in files_to_download:
        if Path(file).exists():
            files.download(file)
            print(f" {file} baixado")
            downloaded += 1
        else:
            print(f"  {file} não encontrado")

    print(f"\n {downloaded} arquivos baixados com sucesso!")

except Exception as e:
    print(f"  Download não disponível: {e}")

# 10) RELATÓRIO FINAL
print(f"\n{'='*70}")
print(" PROJETO CONCLUÍDO!")
print(f"{'='*70}")

if 'best_model_name' in locals():
    print(f" MELHOR MODELO: {best_model_name}")
    print(f" PERFORMANCE: {best_metric['auc']:.4f} AUC")

    print(f"\n RESUMO:")
    print(f"  • Dataset: {X.shape[0]} amostras, {X.shape[1]} features")
    print(f"  • Target: {y.mean():.1%} de atrasos")
    print(f"  • Melhor modelo: {best_model_name} (AUC: {best_metric['auc']:.4f})")

    print(f"\n O QUE IREI FAZER:")
    print("  1. Refinar features e hiperparâmetros")
    print("  2. Implementar validação cruzada")
    print("  3. Desenvolver dashboard de monitoramento")
    print("  4. Criar API para predições em tempo real")

    # Avaliação qualitativa
    if best_metric['auc'] > 0.8:
        print("   Excelente performance! Modelo pronto para produção.")
    elif best_metric['auc'] > 0.7:
        print("   Boa performance. Pode ser usado com monitoramento.")
    else:
        print("   Performance moderada. Recomenda-se melhorias.")

else:
    print(" Projeto não concluído devido a erros.")

print(f"\n {pd.Timestamp.now()}")
print(" Processo finalizado!")
```
